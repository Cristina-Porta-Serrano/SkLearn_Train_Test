We are working with the delayed flights dataset, containing up to 1.936.758 different internal flights in the US for 2008 and their causes for delay, diversion and cancellation; if any. The data comes from the U.S. Department of Transportation's (DOT) Bureau of Transportation Statistics (BTS).

The features of the data set are the following:

**Year** 2008


**Month** 1 (January) - 12 (December)


**DayofMonth** 1-31


**DayOfWeek** 1 (Monday) - 7 (Sunday)


**DepTime** actual departure time (local, hhmm)


**CRSDepTime** scheduled departure time (local, hhmm)


**ArrTime** actual arrival time (local, hhmm)


**CRSArrTime** scheduled arrival time (local, hhmm)


**UniqueCarrier** unique carrier code (company code)


**FlightNum** flight number


**TailNum** plane tail number: aircraft registration, unique aircraft identifier


**ActualElapsedTime** total flight duration in minutes


**CRSElapsedTime** scheduled total flight duration in minutes


**AirTime** in minutes


**ArrDelay** arrival delay, in minutes: A flight is counted as "on time" if it operated less than 15 minutes later the scheduled time shown in the carriers' Computerized Reservations Systems (CRS).


**DepDelay** departure delay, in minutes


**Origin** origin IATA airport code


**Dest** destination IATA airport code


**Distance** in miles


**TaxiIn** approaching time to the boarding bridge once the plane has landed (in minutes)


**TaxiOut** approaching time from the boarding bridge to the runway (in minutes)


**Cancelled** was the flight cancelled (1 = yes, 0 = no)


**CancellationCode** reason for cancellation (A = carrier, B = weather, C = NAS, D = security)


**Diverted** was the flight deiverted (1 = yes, 0 = no)


**CarrierDelay** in minutes: Carrier delay is within the control of the air carrier. Examples of occurrences that may determine carrier delay are: aircraft cleaning, aircraft damage, awaiting the arrival of connecting passengers or crew, baggage, bird strike, cargo loading, catering, computer, outage-carrier equipment, crew legality (pilot or attendant rest), damage by hazardous goods, engineering inspection, fueling, handling disabled passengers, late crew, lavatory servicing, maintenance, oversales, potable water servicing, removal of unruly passenger, slow boarding or seating, stowing carry-on baggage, weight and balance delays.


**WeatherDelay** in minutes: Weather delay is caused by extreme or hazardous weather conditions that are forecasted or manifest themselves on point of departure, enroute, or on point of arrival.


**NASDelay** in minutes: Delay that is within the control of the National Airspace System (NAS) may include: non-extreme weather conditions, airport operations, heavy traffic volume, air traffic control, etc.


**SecurityDelay** in minutes: Security delay is caused by evacuation of a terminal or concourse, re-boarding of aircraft because of security breach, inoperative screening equipment and/or long lines in excess of 29 minutes at screening areas.


**LateAircraftDelay** in minutes: Arrival delay at an airport due to the late arrival of the same aircraft at a previous airport. The ripple effect of an earlier delay at downstream airports is referred to as delay propagation.

First of all, we create a sample of the dataset in order to work more comfortably. In this case, we perform a stratified sample with proportional sampling grouped by the Unique Carrier.

```python
flights_UC = pd.DataFrame(delayed_flights.groupby('UniqueCarrier', group_keys=False).apply(lambda x: x.sample(frac=0.2)))
flights_UC
```

```python
flights_UC.columns
```

```python
flights_UC.isnull().sum()
```

<u>**MinMaxScaler**<u>

First of all, we should analyze which are the predictive features that can affect the Arrival Delay, the target variable. In this case, I would say:

-Month


-Day of week


-DepTime

We use MaxMin Scaler. For each value in a feature, MinMaxScaler subtracts the minimum value in the feature and then divides by the range. The range is the difference between the original maximum and original minimum. MinMaxScaler preserves the shape of the original distribution.

```python
normalized_features_mms = ['Month','DayOfWeek','DepTime'] 
flights_UC_copy = flights_UC.copy()
scaler = MinMaxScaler()
flights_UC_copy[normalized_features_mms] = scaler.fit_transform(flights_UC_copy[normalized_features_mms])
flights_UC_copy
```

<u>**RobustScaler**<u>

Then, we use Robust Scaler to normalize other predictive values with lots of outliers.

```python
normalized_features_rs = ['AirTime', 'DepDelay', 'TaxiIn', 'TaxiOut']
flights_UC_copy_rs = flights_UC.copy()
flights_UC_copy_rs[normalized_features_rs] = RobustScaler().fit_transform(flights_UC_copy[normalized_features_rs])
flights_UC_copy_rs
```

In these graphs we see the difference of the original and normalize data.

```python
plt.figure(figsize=(18,30))

plt.subplot(2,1,1)
sns.boxplot(data=flights_UC[normalized_features_rs], orient='h', palette="Greens");
plt.title('Raw data')

plt.subplot(2,1,2)
sns.boxplot(data=flights_UC_copy_rs[normalized_features_rs], orient='h',palette="Greens")
plt.xlim(-3.0, 4)
plt.title('Normalized with RobustScaler')
plt.grid(axis='x')
```

```python
plt.figure(figsize=(18,30))

plt.subplot(2,1,1)
sns.boxplot(data=flights_UC[normalized_features_mms], orient='h', palette="Greens");
plt.title('Raw data')

plt.subplot(2,1,2)
sns.boxplot(data=flights_UC_copy[normalized_features_mms], orient='h',palette="Greens")
plt.xlim(-3.0, 4)
plt.title('Normalized with MinMaxScaler')
plt.grid(axis='x')
```

We concat both dataframes (the one with minmax scaler performed and the other one with robust scaler performed). As we have repeated columns, we add a suffix to the ones we want to remove in order to be able to remove them. Then, we remove the suffix.

```python
delayed_flights_norm = pd.concat([flights_UC_copy, flights_UC_copy_rs], axis=1)
delayed_flights_norm
```

```python
cols=pd.Series(delayed_flights_norm.columns)
for dup in delayed_flights_norm.columns[delayed_flights_norm.columns.duplicated(keep=False)]: 
    cols[delayed_flights_norm.columns.get_loc(dup)] = ([dup + '.' + str(d_idx) 
                                     if d_idx != 0 
                                     else dup 
                                     for d_idx in range(delayed_flights_norm.columns.get_loc(dup).sum())])
delayed_flights_norm.columns=cols
```

```python
delayed_flights_norm.columns
```

```python
delayed_flights_norm.drop(['Month.1', 'DayofMonth.1', 'DayOfWeek.1', 'DepTime.1', 'CRSDepTime.1',
       'ArrTime.1', 'CRSArrTime.1', 'UniqueCarrier.1', 'FlightNum.1',
       'TailNum.1', 'ActualElapsedTime.1', 'CRSElapsedTime.1', 'AirTime',
       'ArrDelay.1', 'DepDelay', 'Origin.1', 'Dest.1', 'Distance.1',
       'TaxiIn', 'TaxiOut', 'Cancelled.1', 'CancellationCode.1',
       'Diverted.1', 'CarrierDelay.1', 'WeatherDelay.1', 'NASDelay.1',
       'SecurityDelay.1', 'LateAircraftDelay.1'], axis=1, inplace=True)

delayed_flights_norm
```

```python
delayed_flights_norm.columns
```

```python
delayed_flights_norm.columns = delayed_flights_norm.columns.str.replace(".1", "")
# display the dataframe
delayed_flights_norm
```

<u>**Polynomial regression**<u>

Finally, we use polynomial regression with some variables that doesn't have linear relationship.

Regression analysis is used to quantify the relationship between one or more explanatory variables and a response variable.

The most common type of regression analysis is simple linear regression, which is used when a predictor variable and a response variable have a linear relationship. However, sometimes the relationship between a predictor variable and a response variable is not linear.

In these cases, it makes sense to use **polynomial regression**, which can explain the nonlinear relationship between the variables.

First of all, we will visualize the relation between the two chosen explanatory variables in order to see if it is useful to perform a polynomial regression.

```python
pol_features = delayed_flights_norm_dum[['DepDelay', 'TaxiIn']]
sns.scatterplot(data=delayed_flights_norm_dum,x='DepDelay', y='TaxiIn')
```

```python
taxiIn_interpolate = pd.DataFrame(delayed_flights_norm_dum['TaxiIn'].interpolate(method="linear", limit_direction ='forward'))
taxiIn_interpolate.rename(columns={"TaxiIn": "TaxiIn_int"}, inplace=True, errors='ignore')
taxiIn_interpolate
```

```python
delayed_flights_preprocessed = pd.concat([delayed_flights_norm_dum, taxiIn_interpolate], axis=1)
delayed_flights_preprocessed.drop(['TaxiIn'], axis=1, inplace=True)
delayed_flights_preprocessed
```

```python
column_tomove = delayed_flights_preprocessed.pop('TaxiIn_int')
  
# insert column
delayed_flights_preprocessed.insert(25, 'TaxiIn_int', column_tomove)

delayed_flights_preprocessed
```

```python
delayed_flights_preprocessed.isna().sum()
```

```python
pol_features = delayed_flights_preprocessed[['DepDelay', 'TaxiIn_int']]
```

```python
polynomial = preprocessing.PolynomialFeatures(degree = 2)
polynomial.fit(pol_features)
PolynomialFeatures()
delayed_flights_preprocessed_pol = polynomial.transform(pol_features)
delayed_flights_preprocessed_pol
```

```python
X = delayed_flights_preprocessed.iloc[:, 24:25].values
y = delayed_flights_preprocessed.iloc[:, 25].values
```

```python
lin = LinearRegression()
 
lin.fit(X,y)
```

```python
poly = PolynomialFeatures(degree = 4)
X_poly = poly.fit_transform(X)
 
poly.fit(X_poly, y)
lin2 = LinearRegression()
lin2.fit(X_poly, y)
```

```python
# Visualising both Polynomial Regression and Linear Regression results

plt.rcParams.update({'font.size': 15, 'figure.figsize': (20, 18)})
plt.subplot(2,2,1)
plt.scatter(X, y, color = 'seagreen')
plt.plot(X, lin.predict(X), color = 'green')
plt.title('Linear Regression')
plt.xlabel('DepDelay')
plt.ylabel('TaxiIn')

plt.subplot(2,2,2)
plt.scatter(X, y, color = 'seagreen')
plt.plot(X, lin2.predict(poly.fit_transform(X)), color = 'green')
plt.title('Polynomial Regression')
plt.xlabel('DepDelay')
plt.ylabel('TaxiIn')

 
plt.show()
```

```python
# Predicting a new result with Linear Regression after converting predict variable to 2D array
pred = 110.0
predarray = np.array([[pred]])
lin.predict(predarray)
```

```python
# Predicting a new result with Polynomial Regression after converting predict variable to 2D array
pred2 = 110.0
pred2array = np.array([[pred2]])
lin2.predict(poly.fit_transform(pred2array))
```
